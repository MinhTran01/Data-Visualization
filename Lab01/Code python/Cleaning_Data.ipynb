## II. Tiền xử lý dữ liệu

### 1. Xóa cột không cần thiết
#### Nhận xét: 
   Cột '#' là số thứ tự không cần thiết khi phân tích

`````python
df = df.drop('#', axis=1)
df
`````

`````python
# Kiểm tra cột dữ liệu
df.columns
`````

### 2. Chuyển kiểu dữ liệu

`````python
df.info()
`````

#### Nhận xét : 
- Trong mô tả dữ liệu trên:
    - Chỉ có cột `NewDeaths` là có kiểu dữ liệu `float64`
    - Các cột còn lại có kiểu là `object`
- Để phân tích và thống dữ liệu ta cần chuyển kiểu dữ liệu của các cột như đã nhận xét trong phần Tổng quan như sau:
    - Cột `Country,Other` và `Continent` có kiểu dữ liệu là `string`
    - Các cột còn lại có kiểu là `integer`

`````python
df["Country,Other"] = df["Country,Other"].astype('string')
df["Continent"] = df["Continent"].astype('string')
`````

`````python
col = ["Country,Other", "Continent"]

for c in df.columns:
    if c in col:
        continue
    else:
        df[c] = df[c].astype(str)
        df[c] = df[c].str.replace(',', '')
        df[c] = pd.to_numeric(df[c], errors='coerce').astype('Int64') 
`````

`````python
# Kiểm tra lại kiểu dữ liệu
df.info()
`````

### 3. Kiểm tra và xử lý missing value

#### Số lượng missing value trong mỗi cột thuộc tính và tỉ lệ phần trăm của giá trị missing

`````python
count_missing = df.isnull().sum()
percent_missing = round(df.isnull().sum() * 100 / len(df),3)
df_missing = count_missing.to_frame(name='Mumbers')
df_missing.insert(loc=1, column='Percent missing', value=percent_missing)
df_missing
`````

##### Nhận xét
   - Phần lớn các cột đều có giá trị null. Trong đó:
   
       - Các cột có tỉ lệ missing value cao hơn 70% là:
            - `NewCases` :           80,087%
            - `NewDeaths` :          90,043%
            - `NewRecovered` :      78.355%
            
       - Các cột có tỉ lệ missing value trung bình là:
            - `Serious,Critical` :  45.022%
            
       - Các cột có tỉ lệ missing value thấp dưới 10% là:
            - `TotalRecovered` :     9.091%
            - `ActiveCases` :        8.225%
            - `Deaths/1M pop` :      3.463%
            - `TotalTests` :         7.792%
            - `Tests/1M pop` :       7.792%
            - `Continent` :          0.866%
            - `Tot Cases/1M pop` :   0.866%
            
       - Các cột không có missing value là: 
           - ````Country,Other````, ````TotalCases````, ````TotalDeaths````, ````Population````	

#### Xử lý missing value
 * Đối với các cột dữ liệu có tỉ lệ missing value cao hơn 70%
     - Dữ liệu missing ở các cột này khá cao, đồng thời những dữ liệu này cho biết số ca nhiễm, số ca tử vong, số ca hồi phục mới trong ngày và những số liệu này không đóng góp nhiều vào quá trình phân tích và thống kê --> Chọn loại bỏ các cột dữ liệu này
        

`````python
df = df.drop(columns=["NewCases", "NewRecovered", "NewDeaths"])
df
`````

#### Xử lý missing value
 * Đối với các cột dữ liệu có tỉ lệ missing value trung bình

`````python
country_Serious = []
for i in range(0, len(df["Serious,Critical"])):
    if df.isnull().iloc[i, 5]:
        country_Serious.append(df.loc[i, "Country,Other"])
print(country_Serious)
`````

- Nhận xét:
    - Vì số lượng missing value trong cột Serious,Critical cũng ở mức cao, nhưng cột dữ liệu này có ý nghĩa thống kê quan trọng vì vậy nhóm chọn phương pháp điền giá trị trống trong cột này là 0 để mục đích khi phân tích có thể so sánh được mức độ Serious,Critical của các quốc gia 

`````python
df["Serious,Critical"] = df["Serious,Critical"].fillna(0)
df
`````

#### Xử lý missing value
 * Đối với các cột dữ liệu có tỉ lệ missing value thấp:
     - Nhận thấy đối với 2 cặp cột dữ liệu [`Continent`,`Tot Cases/1M pop`] và [`TotalTests` ; `Tests/1M pop`] có tỉ lệ missing value bằng nhau lần lượt là 0.866% và 7.792%
     - Đặt giả thuyết rằng các giá trị missing tại các cặp cột này là của cùng một quốc gia. Ta kiểm tra như sau: 

`````python
country_Continent = []

for i in df.index:

    try:
        if df.isnull().iloc[i, 11]:
            country_Continent.append(i)
            print('Continent _', i, ': ', df.loc[i, "Country,Other"])
    except:
        None
    
    try:
        if df.isnull().iloc[i, 6]:
            print('Tot Cases/1M pop _', i, ': ', df.loc[i, "Country,Other"])
    except:
        None
`````

##### Nhận xét:
- Đúng với giả thuyết các ô dữ liệu bị thiếu thuộc cùng một Quốc gia
    - Tại bộ [`Continent`,`Tot Cases/1M pop`] cột dữ liệu `Continent` xác định Châu lục mà quốc gia đó thuộc về bị null, điều này có thể gây ảnh hưởng đến việc phân tích và thống kê nếu điền giá trị null bằng phương pháp thay thế. Vì vậy nhóm lựa chọn phương án loại bỏ các sample này và xem chúng như outlier của bộ dữ liệu

`````python
df = df.drop(country_Continent, axis=0).reset_index(drop=True)
print("Số sample và thuộc tính còn lại:", df.shape)
`````

`````python
country_TotalTests = []
country_Tests1Mpop = []
for i in df.index:

    try:
        if df.isnull().iloc[i, 8]:
            country_TotalTests.append(i)
            print('TotalTests _', i, ': ', df.loc[i, "Country,Other"], ': ',  df.loc[i, "Population"])
    except:
        None
    
    try:
        if df.isnull().iloc[i, 9]:
            print('Tests1Mpop _', i, ': ', df.loc[i, "Country,Other"], ': ',  df.loc[i, "Population"])
    except:
        None
`````

##### Nhận xét:
- Đúng với giả thuyết các ô dữ liệu bị thiếu thuộc cùng một Quốc gia
    - Tại bộ [`TotalTests` ; `Tests/1M pop`] , Các sample bị thiếu thuộc các quốc gia có số lượng dân số thấp, và nhận thấy rằng nếu loại bỏ các sample này sẽ không làm ảnh hưởng quá nhiều đến việc thống kê sau này, bên cạnh đó còn hạn chế được các giá trị missing đáng kể

`````python
df = df.drop(country_TotalTests, axis=0).reset_index(drop=True)
print("Số sample và thuộc tính còn lại:", df.shape)
`````

#### Kiểm tra lại phần trăm missing value

`````python
count_missing = df.isnull().sum()
percent_missing = round(df.isnull().sum() * 100 / len(df),3)
df_missing = count_missing.to_frame(name='Mumbers')
df_missing.insert(loc=1, column='Percent missing', value=percent_missing)
df_missing
`````

##### Nhận xét:
- Các cột còn lại có missing value khá thấp, nhóm lựa chọn phương pháp giữ nguyên các dữ liệu đó để tránh làm sai lệch dữ liệu quá nhiều
